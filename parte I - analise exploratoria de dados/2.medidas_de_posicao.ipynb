## Medidas de Posição

Usadas para resumir os dados de forma representativa

### Moda

Realização mais frequente do conjunto de valores observados

### Mediana

Realização que ocupa a posição central da série de observações quando estão ordenadas em ordem crescente

### Média Aritmética

É a soma das observações dividida
pela quantidade delas

Para as variáveis nominais somente podemos trabalhar com a moda. Para as variáveis ordinais,
além da moda, podemos usar também a mediana

## Medidas de Dispersão

Medidas que sumarizam a variabilidade de um conjunto de observações e que permite, por exemplo, comparar conjuntos diferentes de valores segundo algum critério estabelecido


### Variância e Desvio Padrão

Tanto a variância como o desvio médio são medidas de dispersão calculadas em relação à
média das observações.

## Quantis Empíricos

Medidas estatísticas que dividem o conjunto de dados em uma determinada proporção. Um exemplo de um quantil é a Mediana, sendo considerada o quantil (50%) ou q(50), pois ela divide a distribuição dos dados em 50%

## Boxplots

O box plot dá uma ideia da posição, dispersão, assimetria, caudas e dados discrepantes

### Outlier

Considerado ponto exterior ao limite do Box Plot, ou seja, valores atípicos

## Gráficos de Simetria

Se os quantis da direita estão mais afastados da mediana, do que os da esquerda, os dados serão assimétricos à direita. Se ocorrer o contrário, os dados serão assimétricos à esquerda


## Transformações

Vários procedimentos estatísticos são baseados na suposição de que os dados provêm de uma
distribuição normal. Mas, em muitas
situações de interesse prático, a distribuição dos dados da amostra é assimétrica e pode conter
valores atípicos. Se quisermos utilizar tais procedimentos com distribuição simétrica, o que se propõe é efetuar uma transformação das
observações, de modo a se obter uma distribuição mais próxima da normal. 

1) Log-transformação: Ideal para dados assimétricos positivos com valores altos.
2) Raiz quadrada: Suaviza assimetrias leves e reduz a escala.
3) Box-Cox: Ajusta os dados para aproximá-los de uma distribuição normal.

#### Moda, Média e Mediana

```python
moda_carros = st.mode(df['Carro'])
moda_carros
```

```python
moda_estados = st.mode(df['Estado'])
moda_estados
```

```python
df['KM'] = pd.to_numeric(df['KM'])

mediana_km = st.median(df['KM'])
media_km = st.mean(df['KM'])

print(f'A mediana é {mediana_km} e a média é {media_km}')
```

```python
df['Valor'] = pd.to_numeric(df['Valor'])

mediana_valor = st.median(df['Valor'])
media_valor = st.mean(df['Valor'])

print(f'A mediana é {mediana_valor} e a média é {media_valor}')
```

```python
df['Viajantes'] = pd.to_numeric(df['Viajantes'])

mediana_viajantes = st.median(df['Viajantes'])
media_viajantes = st.mean(df['Viajantes'])

print(f'A mediana é {mediana_viajantes} e a média é {media_viajantes}')
```

#### Variância e Desvio Padrão


```python
desvio_padrao_km = st.stdev(df['KM'])
variancia_km = st.variance(df['KM'])

desvio_padrao_valor = st.stdev(df['Valor'])
variancia_valor = st.variance(df['Valor'])

desvio_padrao_viajantes = st.stdev(df['Viajantes'])
variancia_viajantes = st.variance(df['Viajantes'])


print(f'Desvio padrão de KM: {desvio_padrao_km}')
print(f'Variância de KM: {variancia_km}')

print(f'Desvio padrão de Valor: {desvio_padrao_valor}')
print(f'Variância de Valor: {variancia_valor}')

print(f'Desvio padrão de Viajantes: {desvio_padrao_viajantes}')
print(f'Variância de Viajantes: {variancia_viajantes}')
```

#### Gráficos de Simetria

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(15, 10))

for i, col in enumerate(['KM', 'Valor', 'Viajantes']):
    plt.subplot(3, 1, i + 1)
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f"Q-Q Plot de {col}", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
