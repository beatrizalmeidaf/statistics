# Variáveis Aleatórias Contínuas

Variável aleatória é um conceito utilizado na teoria das probabilidades e estatística para descrever um tipo de função matemática que relaciona eventos aleatórios a possíveis valores que pertencem a um intervalo de números reais. 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### Distribuição Uniforme

```python
x = np.linspace(0, 1, 1000) # retorna mil números uniformemente espaçados em um intervalo especificado
distribuicao_uniforme = stats.uniform.pdf(x, 0, 1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, distribuicao_uniforme, label='Uniforme(0,1)', color='blue')
plt.fill_between(x, distribuicao_uniforme, alpha=0.2, color='blue')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.title('Distribuição Uniforme')
plt.legend()
plt.grid()
plt.show()
```

#### Distribuição Normal

```python
x = np.linspace(-4, 4, 1000)
distribuicao_normal = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(6, 4))
plt.plot(x, distribuicao_normal, label='Normal(0,1)', color='red')
plt.fill_between(x, distribuicao_normal, alpha=0.2, color='red')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.title('Distribuição Normal')
plt.legend()
plt.grid()
plt.show()
```

#### Distribuição Exponencial

```python
x = np.linspace(0, 5, 1000)
distribuicao_exponencial = stats.expon.pdf(x, 0, 1)

plt.figure(figsize=(6, 4))
plt.plot(x, distribuicao_exponencial, label='Exponencial(λ=1)', color='green')
plt.fill_between(x, distribuicao_exponencial, alpha=0.2, color='green')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.title('Distribuição Exponencial')
plt.legend()
plt.grid()
plt.show()
```

#### Gerando Amostras

```python
samples_normal = np.random.normal(0, 1, 10000)
plt.figure(figsize=(6, 4))
plt.hist(samples_normal, bins=30, density=True, alpha=0.6, color='purple', label='Histograma')
plt.plot(x, distribuicao_normal, 'r-', label='Distribuicao teórica')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.title('Amostras da Distribuição Normal')
plt.legend()
plt.grid()
plt.show()
```
